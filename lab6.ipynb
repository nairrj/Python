Q2. Using a pivot table, find the median flight time to go from each origin to each destination. Save that resulting pivot table in a variable called "pivoted_df".


```python
pivoted_df = flights.pivot_table(index='origin',columns='dest',values='air_time', aggfunc='mean')
```

Q3. Experiment with the .dropna method -- consult its documentation using the ?. Use it to drop destinations in your result from #2 where at least one value is NaN. Store this "cleaned" DataFrame in a variable called "clean_df".

```python
clean_df= flights.pivot_table(index='origin',columns='dest',values='air_time', aggfunc='mean').dropna(axis='columns')
clean_df
```
